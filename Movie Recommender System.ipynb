```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import warnings # for filter out warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
## merging two dataframes
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

## What we are keeping as columns:
1. genres
2. id
3. keywords
4. title
5. overview
6. cast
7. crew

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast # To convert string of list into list.
```

```python
# Function to fetch the genres of movies
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True) # To remove the NA values
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# Function to fetch 3 actors for movies.
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
# function to fetch the name of director of movies.
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
# To remove spacing from the entities of different features so as to have better result.
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Creating new columns with name tags which will be concatenation of overview, genres, keywords,cast, crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# dropping extra columns
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 
```

```python
# function to split the text into lemma.
def lemmatize(text):
    y = []
    for i in text.split():
        y.append(lemmatizer.lemmatize(i))
    return " ".join(y)
```

```python
## Lemmatizing the tags to vectorize.
new['tags']=new['tags'].apply(lemmatize)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
# vectorizing the tags column.
vector = tf.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# finding the cosine si,ilarity to check the closeness of two movies.
similarity = cosine_similarity(vector)
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
# Function to recommend movies
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title) 
```

```python
recommend('Batman Begins')
```
